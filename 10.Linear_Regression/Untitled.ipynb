```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = pd.read_csv("Linear_X_Train.csv")
Y = pd.read_csv("Linear_Y_Train.csv")
X_test = pd.read_csv("Linear_X_Test.csv").values
```

```python
plt.style.use("seaborn")
plt.scatter(X,Y)
plt.show()
```

```python
print(X.shape)
```

```python
X = X.values
Y = Y.values
```

```python
#normalisation
u = X.mean()
std = X.std()
X = (X-u)/std
```

```python
plt.style.use("seaborn")
plt.scatter(X,Y)
plt.title("hardwork-performance")
plt.xlabel("hardwork")
plt.ylabel("time")
plt.show()
```

```python
def hypothesis(x,theta):  # theta = [theta1,theta2]
    y_ = theta[0] + theta[1]*x
    return y_
def gradient(X,Y,theta):
    m=X.shape[0]
    grad = np.zeros((2,))
    for i in range(m):
        y_ = hypothesis(X[i],theta)
        y = Y[i]
        x = X[i]
        grad[0] += (y_-y)
        grad[1] += (y_-y)*x
        
    return grad/m

def error(X,Y,theta):
    m = X.shape[0]
    total_error = 0.0
    for i in range(m):
        y_ = hypothesis(X[i],theta)
        total_error+= (y_ - Y[i])**2
    return total_error    

def gradient_Descent(X,Y,lr = 0.1,max_steps =100):
    theta = np.zeros((2,))
    error_list = []
    for i in range(max_steps):
        grad = gradient(X,Y,theta)
        e = error(X,Y,theta)
        error_list.append(e)
        
        theta[0] = theta[0] - lr*grad[0]
        theta[1] = theta[1] - lr*grad[1]
    return theta,error_list

    
```

```python
theta,error_list = gradient_Descent(X,Y)
```
